```python
tipo_fundo = spark_session.createDataFrame([
        ('280', '2019-03-27', None),
        ('256', '2019-03-27', '2019-10-31'),
        ('193', '2019-03-27', '2019-10-04'),
        ('273', '2019-11-03',  None),
        ('311', '2019-11-01',  None),
        ('312', '2019-11-01',  None),
        ('400', '2019-11-01',  None),
    ],["tipo_fundo_codigo", "tipo_fundo_data_inicio","tipo_fundo_data_fim"]) 
```

```python
def fundos_encerrados_novo_tipox(fato_movimento_transferencia_final, dim_tipo_fundo):
    fato_movimento_transferencia_final = fato_movimento_transferencia_final.alias('l').join( broadcast(dim_tipo_fundo).alias('r'),
        col('l.codtipo_anterior') == col('r.tipo_fundo_codigo') , 'left' )
    
    fato_movimento_transferencia_final = fato_movimento_transferencia_final.withColumn("aux_column",
        when(
                ( col('tipo_fundo_data_fim').isNotNull() ) & ( col('codtipo') != col('codtipo_anterior') )  
                , col("codtipo")
            ).otherwise(None)
    )
    window = Window.partitionBy( ["codfundo"] ).orderBy( desc('data') )
       
    
    fato_movimento_transferencia_final = fato_movimento_transferencia_final.withColumn("novo_codtipo",
                    F.first('aux_column', True).over(window))

    fato_movimento_transferencia_final = fato_movimento_transferencia_final.withColumn("test", 
            F.collect_list('aux_column').over(window))
    
    fato_movimento_transferencia_final = fato_movimento_transferencia_final\
        .withColumn( 'novo_codtipo', when( col('novo_codtipo').isNull(), 
            col('codtipo') ).otherwise( col('novo_codtipo') ) )                                                                                
    
    return fato_movimento_transferencia_final.drop('tipo_fundo_codigo') 
```

```python
fundos_encerrados_novo_tipox(novo_tipo, tipo_fundo).orderBy('codfundo', col('data').asc()).toPandas()
```
