```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
from pyspark.sql import functions as F, types as T
import datetime
data_filter = datetime.date(2015, 10, 1)

data_1 = datetime.date(2015, 9, 30)
data_2 = datetime.date(2015, 10, 1)
data_3 = datetime.date(2015, 10, 2)
data_4 = datetime.date(2017, 12, 21)

# fundo_dia = spark.createDataFrame([(data_1, '486371', '238'),
#                                   (data_2, '486371', '238'),
#                                   (data_3, '486371', '238'),
#                                   (data_4, '486371', '238')
#                                   ],['data', 'codfundo', 'codtipo'] )
```

```python
fundo_dia = spark.createDataFrame([
                                    (data_2, '486371', '238'),
                                  ],['data', 'codfundo', 'codtipo'] )
```

```python
dim_categoria_fundo_aux_df = spark.read\
    .parquet("dim_categoria_fundo_aux").filter(F.col("categoria_codigo_fundo") == '486371')
dim_categoria_fundo_aux_df = dim_categoria_fundo_aux_df\
    .groupBy(['categoria_codigo_fundo',
              'categoria_tipo_descricao',
              'categoria_codigo_si',
              'categoria',
              'indicador_atual',
              'categoria_data_inicio',
              'categoria_data_fim'])\
    .pivot("categoria_codigo_si").agg(F.first("categoria"))
dim_categoria_fundo_aux_df = dim_categoria_fundo_aux_df.drop('categoria')
```

```python
check_columns = [str(i) for i in range(1, 22)]
for column in check_columns:
    if column not in dim_categoria_fundo_aux_df.columns:
        dim_categoria_fundo_aux_df = dim_categoria_fundo_aux_df.withColumn(
            column, F.lit(None))
```

```python
dim_categoria_fundo_aux_df = dim_categoria_fundo_aux_df\
    .withColumnRenamed("1", "orgao_regulador")\
    .withColumnRenamed("2", "status_fundo")\
    .withColumnRenamed("3", "previdencia")\
    .withColumnRenamed("4", "alavancado")\
    .withColumnRenamed("5", "tipo_investidor")\
    .withColumnRenamed("6", "longo_prazo_cvm")\
    .withColumnRenamed("7", "tributacao_alvo")\
    .withColumnRenamed("8", "classe_cvm")\
    .withColumnRenamed("9", "credito_privado")\
    .withColumnRenamed("10", "invest_no_exterior")\
    .withColumnRenamed("11", "resp_limites_e_a")\
    .withColumnRenamed("12", "aplicacao_automatica")\
    .withColumnRenamed("13", "foco_atuacao")\
    .withColumnRenamed("14", "previdenciario")\
    .withColumnRenamed("15", "sub_classe_cvm")\
    .withColumnRenamed("16", "classe_anbima")\
    .withColumnRenamed("17", "categoria")\
    .withColumnRenamed("18", "sub_categoria")\
    .withColumnRenamed("19", "benchmark")\
    .withColumnRenamed("20", "multigestor")\
    .withColumnRenamed("21", "invest_qualificado")

dim_categoria_fundo_aux_df = dim_categoria_fundo_aux_df.orderBy('categoria_data_inicio')
```

```python
dimension = fundo_dia.alias('fd_pl').join(dim_categoria_fundo_aux_df.alias('cat'),
        (
            fundo_dia.codfundo == dim_categoria_fundo_aux_df.categoria_codigo_fundo
        ) &
        (
            fundo_dia.data.between(
            dim_categoria_fundo_aux_df.categoria_data_inicio,
            dim_categoria_fundo_aux_df.categoria_data_fim)
        ), 'left')
```

```python
dimension.orderBy('categoria_codigo_si')\
.select(['codfundo', 'data', 'codtipo', 'categoria_data_inicio','categoria_data_fim',
        'categoria_codigo_si', 
        'orgao_regulador',
        'previdencia',
        'alavancado',
        'longo_prazo_cvm',
        'tributacao_alvo',
        'classe_cvm',
        'credito_privado',
        'invest_no_exterior',
        'resp_limites_e_a',
        'aplicacao_automatica',
        'previdenciario',
        'sub_classe_cvm',
        'classe_anbima',
        'categoria',
        'sub_categoria',
        'benchmark',
        'multigestor',
        'invest_qualificado',])\
.toPandas()
```

```python
a = dimension.groupBy([
    'data',
    'codfundo',
    'codtipo']) \
    .agg(
    (F.first('orgao_regulador', True)).alias("orgao_regulador"),
    (F.first('status_fundo', True)).alias("status_fundo"),
    (F.first('previdencia', True)).alias("previdencia"),
    (F.first('alavancado', True)).alias("alavancado"),
    (F.first('tipo_investidor', True)).alias("tipo_investidor"),
    (F.first('longo_prazo_cvm', True)).alias("longo_prazo_cvm"),
    (F.first('tributacao_alvo', True)).alias("tributacao_alvo"),
    (F.first('classe_cvm', True)).alias("classe_cvm"),
    (F.first('credito_privado', True)).alias("credito_privado"),
    (F.first('invest_no_exterior', True)).alias("invest_no_exterior"),
    (F.first('resp_limites_e_a', True)).alias("resp_limites_e_a"),
    (F.first('aplicacao_automatica', True)).alias("aplicacao_automatica"),
    (F.first('foco_atuacao', True)).alias("foco_atuacao"),
    (F.first('previdenciario', True)).alias("previdenciario"),
    (F.first('sub_classe_cvm', True)).alias("sub_classe_cvm"),
    (F.first('classe_anbima', True)).alias("classe_anbima"),
    (F.first('categoria', True)).alias("categoria"),
    (F.first('sub_categoria', True)).alias("sub_categoria"),
    (F.first('benchmark', True)).alias("benchmark"),
    (F.first('multigestor', True)).alias("multigestor"),
    (F.first('invest_qualificado', True)).alias("invest_qualificado")
)
```
